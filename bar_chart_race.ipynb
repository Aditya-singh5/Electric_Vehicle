```
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```
df = pd.read_csv('dataset.csv')
df.head()
```

```
import bar_chart_race as bcr
```

```
df_number = df.groupby(['Model Year','Make']).size().reset_index(name='Vehicle_Number')
```

```
df_pivot = pd.pivot_table(df_number,index='Model Year', columns='Make', values='Vehicle_Number')

print(df_pivot)
```
